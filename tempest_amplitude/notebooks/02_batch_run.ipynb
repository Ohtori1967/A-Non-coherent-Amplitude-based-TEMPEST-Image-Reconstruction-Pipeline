### 02: Module Import

### 03: Load Configuration

### 04: Run Batch Process

```python
df = batch_process(
    input_dir=cfg["input"]["input_dir"],
    output_dir=cfg["output"]["output_dir"],
    cfg=cfg,
    pattern=cfg["input"]["pattern"],
    recursive=cfg["input"]["recursive"],
    show_progress=True,
)
```

### 05: Detect Results

```python
df.head()
```

### 06: Results Statistics

```python
if not df.empty:
    print("Total   :", len(df))
    print("Success :", int(df["success"].sum()))
    print("Fail    :", int((~df["success"]).sum()))

    valid_foff = df.loc[df["success"], "freq_offset_hz"].dropna()
    if len(valid_foff) > 0:
        print("Mean freq offset (Hz):", float(valid_foff.mean()))
```
