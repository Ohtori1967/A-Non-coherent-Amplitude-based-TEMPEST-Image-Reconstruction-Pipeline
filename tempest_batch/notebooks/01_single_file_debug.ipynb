#### Inspect Result Summary

#### Preview Saved Files

#### Quick Experiment Tips
- Set `frame_count = 1` for single-frame debugging.
- Increase `frame_count` to inspect stacked output.
- Adjust `start_offset_samples` to test frame alignment.
- Use `image_mode = "log_gain"` for cleaner stacked output.
