### 02 Batch Run
Run reconstruction for multiple captured signal batches.

```python
import sys
from pathlib import Path
     
project_root = Path("..").resolve()
sys.path.append(str(project_root / "src"))
```

#### Imports

```python
import pandas as pd
from IPython.display import display

from recon.config_manager import load_config
from recon.batch_runner import batch_process_collection
```

#### Load Configuration

```python
cfg = load_config(project_root / "configs" / "recon" / "batch_default.json")

# Optional overrides for this notebook run
# cfg["input"]["input_root_dir"] = r"E:/UsersData/Desktop/usrp_signal_capture/outputs"
# cfg["output"]["output_dir"] = str(project_root / "outputs")
# cfg["stacking"]["frame_count"] = 4
# cfg["stacking"]["frame_start"] = 0
# cfg["stacking"]["image_mode"] = "log_gain"
# cfg["processing"]["start_offset_samples"] = 0

cfg
```

#### Show Key Parameters

```python
print("===== Input =====")
print("input_root_dir :", cfg["input"]["input_root_dir"])
print("pattern        :", cfg["input"]["pattern"])
print("recursive      :", cfg["input"]["recursive"])
print("exclude_dirs   :", cfg["input"]["exclude_dir_names"])

print("\n===== Output =====")
print("output_root    :", cfg["output"]["output_dir"])
print("save_iq_png    :", cfg["output"]["save_iq_png"])
print("save_amp_png   :", cfg["output"]["save_amp_png"])
print("save_amp_gain  :", cfg["output"]["save_amp_gain_png"])
print("save_amp_log   :", cfg["output"]["save_amp_log_gain_png"])
print("save_meta_json :", cfg["output"]["save_meta_json"])

print("\n===== Signal =====")
print("Fs             :", cfg["signal"]["Fs"])
print("pixel_rate     :", cfg["signal"]["pixel_rate"])
print("h_total        :", cfg["signal"]["h_total"])
print("v_total        :", cfg["signal"]["v_total"])
print("up/down        :", cfg["signal"]["up"], "/", cfg["signal"]["down"])

print("\n===== Processing =====")
print("BW             :", cfg["processing"]["BW"])
print("numtaps        :", cfg["processing"]["numtaps"])
print("frame_idx      :", cfg["processing"]["frame_idx"])
print("offset         :", cfg["processing"]["start_offset_samples"])
print("remove_dc      :", cfg["processing"]["remove_dc"])
print("foff_corr      :", cfg["processing"]["freq_offset_correction"])

print("\n===== Stacking =====")
print("frame_start    :", cfg["stacking"]["frame_start"])
print("frame_count    :", cfg["stacking"]["frame_count"])
print("direction      :", cfg["stacking"]["direction"])
print("image_mode     :", cfg["stacking"]["image_mode"])
print("normalization  :", cfg["stacking"]["normalization"])
```

#### Run Batch Reconstruction

```python
# Unified mode:
# use only "stack"
# - frame_count = 1 -> single-frame special case
# - frame_count > 1 -> stacked multi-frame output
run_mode = "stack"

all_results = batch_process_collection(
    input_root_dir=cfg["input"]["input_root_dir"],
    output_root_dir=cfg["output"]["output_dir"],
    cfg=cfg,
    pattern=cfg["input"]["pattern"],
    recursive=cfg["input"]["recursive"],
    show_progress=True,
    mode=run_mode,
)

print(f"Processed batch folders: {len(all_results)}")
list(all_results.keys())[:10]
```

#### Preview One Batch

```python
if len(all_results) == 0:
    print("No batch results.")
else:
    first_batch_name = next(iter(all_results))
    print("First batch:", first_batch_name)
    display(all_results[first_batch_name].head())
```

#### Load Collection Summary

```python
collection_csv = Path(cfg["output"]["output_dir"]) / "collection_summary.csv"
collection_df = pd.read_csv(collection_csv, encoding="utf-8-sig")

print("Collection summary:", collection_csv)
display(collection_df.head())
```

#### Key Columns Preview

```python
cols_to_show = [
    "stem",
    "success",
    "run_mode",
    "input_batch_name",
    "loaded_samples",
    "resampled_samples",
    "available_frames",
    "freq_offset_hz",
    "Fs",
    "Fc",
    "capture_time_s",
    "total_overrun",
    "capture_meta.sample_id",
    "capture_meta.slide_index",
    "capture_meta.font_size_pt",
    "capture_meta.sdr_model",
    "capture_meta.antenna_model",
    "capture_meta.test_distance_cm",
    "stack_frame_start",
    "stack_frame_count",
    "stack_direction",
    "stack_image_mode",
]

existing_cols = [c for c in cols_to_show if c in collection_df.columns]
display(collection_df[existing_cols].head(20))
```

#### Basic Statistics

```python
print("Total files   :", len(collection_df))
print("Success files :", int(collection_df["success"].sum()))
print("Failed files  :", int((~collection_df["success"]).sum()))

valid_foff = collection_df.loc[collection_df["success"] == True, "freq_offset_hz"].dropna()
if len(valid_foff) > 0:
    print("Mean freq offset (Hz):", float(valid_foff.mean()))
    print("Std  freq offset (Hz):", float(valid_foff.std()))
```

#### Show Batch Names

```python
if "input_batch_name" in collection_df.columns:
    display(collection_df["input_batch_name"].dropna().drop_duplicates().sort_values().to_frame(name="batch_name"))
else:
    print("No input_batch_name column found.")
```

#### Preview Failed Rows

```python
fail_df = collection_df.loc[collection_df["success"] == False].copy()
print("Failed rows:", len(fail_df))
display(fail_df.head(20))
```
